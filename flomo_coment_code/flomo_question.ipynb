```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_new = pd.read_excel('output2.xlsx')
data_new.head(1)
```

```python
data_filter = data_new[data_new['type'].isin(['问题反馈','痛点/抱怨'])]
data_filter.head(1)
```

```python
data_filter.shape
```

```python
columns_needs = ['note_id','content','topic','like_count','sub_comment_count']
data_filter_selected = data_filter[columns_needs]
sorted_df = data_filter_selected.sort_values(['like_count','sub_comment_count'],ascending = False)
sorted_df.head(50)
```

```python
sorted_df.to_excel('痛点.xlsx',index = False)
```

```python
### flomo 价格
data_filter2 = data_new[data_new['topic'].isin(['价格'])]
data_filter2.head(1)
data_filter2.shape
```

```python
columns_needs = ['note_id','content','topic','like_count','sub_comment_count']
data_filter2_selected = data_filter2[columns_needs]
sorted_df = data_filter2_selected.sort_values(['like_count','sub_comment_count'],ascending = False)
sorted_df.head(48)
```
